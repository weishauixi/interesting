```python
arr=np.random.randint(1,20,8)
print(arr)
print(np.sum(arr))

#计算平均值
print(np.mean(arr))
#计算中位数
print(np.median([1,2,4]))
#方差
print(np.var([1,2,3]))
print(np.std([1,2,3]))
#分位数
np.random.seed(0)
arr=np.random.randint(1,20,8)
print(arr)
print(np.median(arr))
```

```python
#比较函数
print(np.greater([3,4,5,6,7],4))
print(np.less([3,4,5,6,7],4))
print(np.equal([3,4,5,6,7],4))
print(np.equal([3,4,5,6,7],[3,4,5,6,7]))
```

```python
#自定义条件
#print(np.where(条件，符合条件，不符合条件))
arr=np.array([1,2,3,4,5])
print(np.where(arr<3,1,0))
```

```python
score=np.random.randint(50,100,20)

print(score)
print(np.where(score>50,1,0))

print(np.where(
    score<60,'不及格',np.where(
        score<80,'良好','优秀'
    )))
```

```python
#np.select(条件，返回的结果)
print(np.select([score>=80,(score>=60) & (score<80),score<60],['优秀','良好','不及格'],default='为之'))
```

```python
'''
折线图
'''
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

import seaborn as sns
plt.figure(figsize=(10,5))
month=['1月','2月','3月','4月']
sales=[150,3,4,435]
plt.plot(month,sales
         ,label='产品A',
         color='orange',
         linewidth=2,
         linestyle='--',
         marker='o',)
plt.title('',color='white',fontsize=10)
#添加坐标轴标签
plt.xlabel('',fontsize=10)
plt.ylabel('',fontsize=10)
#图例
plt.legend(loc='upper left')
# #添加网格线
plt.grid(True,alpha=0.5,color='yellow',linestyle='--')
# plt.grid(axis='x',)
#设置刻度字体
plt.xticks(rotation=60,fontsize=10)
plt.yticks(rotation=0,fontsize=10)
#设置默认的刻度
plt.ylim(0,160)
#数据段上显示数值
for x,y in zip(month,sales):
    print(x,y)
    plt.text(x,y,str(y),ha='center',va='bottom',fontsize=10)
plt.show()
```
